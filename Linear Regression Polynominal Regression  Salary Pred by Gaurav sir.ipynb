# Salary Prediction  With Acurracy find out karo 

```python
x= np.array([[2],[3],[4],[6],[7],[10]])
y= np.array([20000,25000,35000,50000,55000,75000])
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(x,y)
```

```python
m = lr.coef_
m
```

```python
c = lr.intercept_
c
```

```python
x = int(input("Enter the Expirence"))
y = m*x+c

y
```

# POLOYNOMINAL REGRESSION 
# agar hamara data point Linear Regression se fit nahi hota tab ham Polynominal se FIt karenge 

```python
# y=m0+m1x+m2x^2+m3x^3------mnx^n
import numpy 

x1 = numpy.array([2,3,4,6,7,10])
y1 = numpy.array([20000,25000,35000,50000,55000,75000])
```

```python
mymodel = numpy.poly1d(numpy.polyfit(x1,y1,3))

#1 degree pe Linear Regresssion rehta hai
```

```python
pred=mymodel(x1)
pred
```

```python
import matplotlib.pyplot as plt
plt.scatter(x1,y1)
plt.plot(x1,pred)
plt.show()


#""""prediction line khicho line
#maen square Error 
#Loss data """"""
```

```python
from sklearn.metrics import r2_score
print(r2_score(y1,pred))
```
